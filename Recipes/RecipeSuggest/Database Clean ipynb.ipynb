```python
# write a function to convert a thing to list and get rid of duplicates and nan values and weird \xa0 thing


def makeList(stuff):
    newList = stuff.unique().tolist()
    newList = [x for x in newList if str(x) != 'nan']
    newList = [x.replace('\xa0', ' ') for x in newList]
    return newList
    
# write a function to add the entries to make them into a class



```

# First find all the unique words in all recipes (to remove ones that dont need to be there)

```python
unique = {}
for value in ingred:
    for i in value:
        row = i.split(' ')
        for word in row:
            new_word = ''
            for char in word:
                if char.isalpha():
                    new_word += char.lower()
            
            if new_word in unique:
                unique[new_word] += 1
            else:
                unique[new_word] = 1
            
```

```python
total_words = 0
over_200 = 0
collection = []
over_200_cnt = 0
for keys, values in unique.items():
    total_words += values
    if values > 200:
        over_200_cnt += values
        over_200 += 1
        collection.append(keys)

print('Percentage of words over 200')
print(over_200_cnt * 100 / total_words)

print('Number of word types over 200')
print(over_200)
```

```python
# so if ingredients word contains word to remove it should be removed
ingred = dff.ingredients
words_to_remove

Glossary= []
for row in ingred:
    glossary = []
    for ingredient in row:
        
        words = ingredient.split(' ')
        for word2 in words:
            new_word = ''
            for char in word2:
                if char.isalpha():
                    new_word += char.lower()
            
            if new_word not in words_to_remove:
                glossary.append(new_word)
    Glossary.append(glossary)
        
dff = dff.assign(Glossary=Glossary)
```
